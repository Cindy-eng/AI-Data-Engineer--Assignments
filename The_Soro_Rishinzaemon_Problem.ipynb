```python
#Problem 1: Number of rice grains on day 100
import matplotlib.pyplot as plt

grains = 1
total_grains = grains
rice_per_day = [grains]

# Loop to calculate grains for 100 days
for i in range(1, 100):
    grains *= 2
    rice_per_day.append(grains)
    total_grains += grains

print(f'The total grains on the 100th day are {rice_per_day}')
print(f'The total grains over 100 days are {total_grains}')

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(range(1, 101), rice_per_day, label='Rice grains received each day', color='blue')
plt.title('Number of Grains of Rice Over 100 Days')
plt.xlabel('Days')
plt.ylabel('Number of Rice Grains (Log Scale)')
plt.yscale('log')
plt.legend()
plt.grid()
plt.show()
```

```python
#Problem 2: Number of rice grains other than the 100th day
import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(days=100):
    list_grains = [1]  # List to store grains received each day
    list_total_grains = [1]  # List to store cumulative total grains

    for i in range(1, days):
        grains = list_grains[-1] * 2
        list_grains.append(grains)
        list_total_grains.append(list_total_grains[-1] + grains)

    return list_grains, list_total_grains

# Use the function to calculate rice grains for a custom number of days
days = 3
list_grains, list_total_grains = compute_sorori_shinzaemon(days=days)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(range(1,days + 1), list_grains, label="Rice received each day", color="blue")
plt.plot(range(1, days + 1), list_total_grains, label="Total rice received by each day", color="orange")
plt.title(f"Change in Rice Grains Over {days} Days")
plt.xlabel("Day")
plt.ylabel("Number of Rice Grains")
plt.yscale("log")  # Logarithmic scale for better visualization
plt.legend()
plt.grid()
plt.show()

# Print total grains on the last day
print(f"The total rice grains received by day {days} are {list_total_grains[-1]}")
```

```python
#Problem 3: How many people can live on the rice they can get and for how many days?
##My investigation
##Rice consumption per person per day:
##The average daily rice consumption per person varies by region, but it's about 400 grams per person per day in many rice-consuming cultures.
##There are approximately 48 grains of rice in 1 gram (source: USDA and estimates from food science resources).
##Therefore, each person consumes
##400×48=19,200 grains of rice per day.
##So my formula is days=Total Grains/(Number of People×19,200)

def calculate_days(quantity_rice , number_people):
    # Average daily rice consumption per person (in grains)
    grains_per_person_per_day = 19200  # 400 grams/day * 48 grains/gram
    # Calculate total daily consumption for the group
    daily_consumption = num_people * grains_per_person_per_day
    # Calculate the number of days the group can survive
    days =quantity_rice  // daily_consumption
    return days

# Total grains from Problem 2 (for exemplo :12 days of rice doubling)
quantity_rice = 560000
num_people = 4

days_survived = calculate_days(quantity_rice, num_people)
print(f"A group of {num_people} people can survive for {days_survived} days on {quantity_rice } grains of rice.")
```
