```python
import matplotlib.pyplot as plt

CHESTNUT_BUN_VOLUME = 0.0001  # in cubic kilometers
SOLAR_SYSTEM_VOLUME = 1.5e18  # in cubic kilometers
DOUBLING_TIME_MINUTES = 5

# Function to calculate time to fill the solar system
def time_to_fill(volume_item, volume_target, doubling_time):
    time_minutes, current_volume = 0, volume_item
    while current_volume < volume_target:
        current_volume *= 2
        time_minutes += doubling_time
    return time_minutes

# Calculate and print time to fill solar system
print(f"Time to fill the solar system: {time_to_fill(CHESTNUT_BUN_VOLUME, SOLAR_SYSTEM_VOLUME, DOUBLING_TIME_MINUTES):.2f}")

# Function to plot growth over time
def plot_growth(volume_item, volume_target, doubling_time):
    times, volumes = [], []
    current_volume, time_minutes = volume_item, 0
    while current_volume < volume_target:
        times.append(time_minutes / 60)  # Convert minutes to hours
        volumes.append(current_volume)
        current_volume *= 2
        time_minutes += doubling_time

    plt.plot(times, volumes, label="Chestnut Bun Volume")
    plt.axhline(y=volume_target, color='r', linestyle='--', label="Target Volume")
    plt.yscale("log")
    plt.xlabel("Time (hours)")
    plt.ylabel("Volume (km^3)")
    plt.title("Growth of Chestnut Buns Over Time")
    plt.legend()
    plt.grid()
    plt.show()

# Plot growth for the solar system
plot_growth(CHESTNUT_BUN_VOLUME, SOLAR_SYSTEM_VOLUME, DOUBLING_TIME_MINUTES)

```
