```python
#[Problem 1] Creating random numbers
import numpy as np
np.random.seed(0)
mean= [-3, 0]
covarianc = [[1.0, 0.8], [0.8, 1.0]]

data_1 = np.random.multivariate_normal(mean, covarianc, 500)
print(data_1)
```

```python
#[Problem 2] Visualization by scatter plot
import matplotlib.pyplot as plt
plt.scatter(data_1[:, 0], data_1[:, 1], alpha=0.6, label="Data 1")
plt.title("Scatter Plot of Data 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#[Problem 3] Visualization by histogram
import matplotlib.pyplot as plt;


plt.subplot(1, 2, 1)
plt.hist(data_1[:, 0], bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title("Histogram of Dimension 1 ")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.xlim(-6, 2)

plt.subplot(1, 2, 2)
plt.hist(data_1[:,1] , bins=30, color='green', edgecolor='black')
plt.title('Histogram of Dimension 2 ')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.xlim(-6, 2)


plt.show();
```

```python
#Problem 4: Adding Data
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
mean1= [0, -3]
covarianc1 = [[1.0, 0.8], [0.8, 1.0]]
data_2 = np.random.multivariate_normal(mean1, covarianc1, 500)

plt.figure(figsize=(8, 6))
plt.scatter(data_1[:, 0], data_1[:, 1], alpha=0.6, label='Data 1', color='blue')
plt.scatter(data_2[:, 0], data_2[:, 1], alpha=0.6, label='Data 2', color='orange')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()

print(data_2)
```

```python
#Problem 5: Combining data
import numpy as np

data_3=np.vstack((data_1,data_2))
data_4=np.concatenate((data_1,data_2))
print('Using np.vstack() :',data_3)
print(' ')
print('Using .np.concatenate(',data_4)
```

```python
#Problem 6: Labeling
labels_1 = np.zeros((500, 1))
labels_2 = np.ones((500, 1))

labeled_data = np.hstack((data_3, np.vstack((labels_1, labels_2))))
print(labeled_data[:10])
```
