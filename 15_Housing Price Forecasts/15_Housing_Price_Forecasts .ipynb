```python
#[Problem 1] Feature selection for practice

import pandas as pd

df=pd.read_csv('train.csv')

expla=df[['GrLivArea','YearBuilt']]
target=df['SalePrice']



```

```python
#[Problem 2] Estimation and evaluation by linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

model = LinearRegression()
x=expla.values
y=target

x_train,x_test,y_train,y_test=train_test_split(x,y);

model.fit(x_train,y_train)
pred=model.predict(x_test)

print(mean_squared_error(y_test,pred))


```

```python
#[Problem 3] Comparison of methods

from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


model1=LinearRegression()
model2=SVR()
model3=DecisionTreeRegressor()
model4=RandomForestRegressor()

x=expla.values
y=target

x_train,x_test,y_train,y_test=train_test_split(x,y)

model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
model4.fit(x_train,y_train)

pred1=model1.predict(x_test)
pred2=model2.predict(x_test)
pred3=model3.predict(x_test)
pred4=model4.predict(x_test)

mean_squared_error(y_test,pred1)
mean_squared_error(y_test,pred2)
mean_squared_error(y_test,pred3)
mean_squared_error(y_test,pred4)



plt.figure(figsize=(20,6))
plt.scatter(y_test,pred1,label='LinearRegression')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', label='Ideal')
plt.grid(True)
plt.show()

plt.scatter(y_test,pred2,label='SVR')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', label='Ideal')
plt.grid(True)
plt.show()

plt.scatter(y_test,pred3,label='DecisionTreeRegressor')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', label='Ideal')
plt.grid(True)
plt.show()


plt.scatter(y_test,pred4,label='RandomForestRegressor')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', label='Ideal')
plt.grid(True)
plt.show()

ms_result={}

ms_result['LinearRegression']=mean_squared_error(y_test,pred1)
ms_result['SVR']=mean_squared_error(y_test,pred2)
ms_result['DecisionTreeRegressor']=mean_squared_error(y_test,pred3)
ms_result['RandomForestRegressor']=mean_squared_error(y_test,pred4)
print(ms_result)
```

```python
#[Problem 4] (Advance assignment) Learning using other features
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
df=pd.read_csv('train.csv')

df_columns = df[['Id','MSSubClass','PoolArea','Neighborhood','GrLivArea','YearBuilt']]
df_target = df['SalePrice']

df_columns_encoded = pd.get_dummies(df_columns, columns=['Neighborhood'])

print(df_columns_encoded.describe())

x = df_columns_encoded.values
y = df_target.values


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)
pred = model.predict(x_test)

print( mean_squared_error(y_test, pred))

plt.figure(figsize=(20,6))
plt.scatter(y_test,pred,label='LinearRegression')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', label='Ideal')
plt.grid(True)
plt.show()

```
