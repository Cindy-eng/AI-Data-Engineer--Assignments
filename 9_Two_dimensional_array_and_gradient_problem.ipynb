```python
#[Problem 1] Linear function
import numpy as np

x=np.arange(-50,50.1,0.1);

y=(1/2)*x+1

print('Y shape:',y.shape,' and X shape',x.shape)
print('x:',x)
print('y:',y)
```

```python
#[Problem 2] Array combination
array_xy = np.column_stack((x, y))

print(array_xy)
print("Shape of array_xy:", array_xy.shape)
```

```python
#[Problem 3] Find the gradient
dy = y[1:] - y[:-1]
dx = x[1:] - x[:-1]

gradient = dy / dx
print("First 10 gradients:", gradient[:10])
print("Gradient shape:", gradient.shape)
```

```python
#[Problem 4] Draw a graph
import matplotlib.pyplot as plt
plt.subplot(1, 2, 2)
plt.plot(x[:-1], gradient, label='Gradient', color='red')
plt.plot(x,y,label='y')
plt.xlabel('x')
plt.ylabel('Gradient')
plt.title('Gradient of the Function')
plt.legend()
plt.grid()
```

```python
#[Problem 5] Python functionalization

def function1(x):
  return x**2;
def function2(x):
  return 2*x**2+2**x;
def function3(x):
  return np.sin(x**(1/2))
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
  x=np.arange(*x_range);
  y=function(x);
  dy = y[1:] - y[:-1]
  dx = x[1:] - x[:-1]

  gradient = dy / dx

  array_xy=np.column_stack((x,y))

  return array_xy ,gradient;

array_x, gradient = compute_gradient(function1);

plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_x[:,0],array_x[:,1],label='y',ls='--')
plt.plot(array_x[:,0][1:],gradient,label='gradient',ls='--')
plt.legend()
plt.show

```

```python

array_x, gradient = compute_gradient(function2);

plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_x[:,0],array_x[:,1],label='y',ls='--')
plt.plot(array_x[:,0][1:],gradient,label='gradient',ls='--')
plt.legend()
plt.show
```

```python
array_x, gradient = compute_gradient(function3);

plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_x[:,0],array_x[:,1],label='y',ls='--')
plt.plot(array_x[:,0][1:],gradient,label='gradient',ls='--')
plt.legend()
plt.show
```

```python
#[Problem 6] Find the minimum value

x_range1 = np.arange(-50, 50.1, 0.1)
x_range2 = np.arange(-50, 50.1, 0.1)
x_range3 = np.arange(0, 50.1, 0.1)

y1 = function1(x_range1)
y2 = function2(x_range2)
y3 = function3(x_range3)

min_index1 = np.argmin(y1)
min_index2 = np.argmin(y2)
min_index3 = np.argmin(y3)

min_y1, min_x1 = y1[min_index1], x_range1[min_index1]
min_y2, min_x2 = y2[min_index2], x_range2[min_index2]
min_y3, min_x3 = y3[min_index3], x_range3[min_index3]

print(f"Function 1: y = x^2  Min y: {min_y1} at x = {min_x1}")
print(f"Function 2: y = 2x^2 + 2x  Min y: {min_y2} at x = {min_x2}")
print(f"Function 3: y = sin(sqrt(x))  Min y: {min_y3} at x = {min_x3}")

def compute_gradient(y_values, x_values):
    return np.diff(y_values) / np.diff(x_values)

gradient1 = compute_gradient(y1, x_range1)
gradient2 = compute_gradient(y2, x_range2)
gradient3 = compute_gradient(y3, x_range3)

if min_index1 > 0 and min_index1 < len(gradient1):
    slope_before1, slope_after1 = gradient1[min_index1-1], gradient1[min_index1]
else:
    slope_before1, slope_after1 = None, None

if min_index2 > 0 and min_index2 < len(gradient2):
    slope_before2, slope_after2 = gradient2[min_index2-1], gradient2[min_index2]
else:
    slope_before2, slope_after2 = None, None

if min_index3 > 0 and min_index3 < len(gradient3):
    slope_before3, slope_after3 = gradient3[min_index3-1], gradient3[min_index3]
else:
    slope_before3, slope_after3 = None, None

print(f"Slope before/after min y for Function 1: {slope_before1}, {slope_after1}")
print(f"Slope before/after min y for Function 2: {slope_before2}, {slope_after2}")
print(f"Slope before/after min y for Function 3: {slope_before3}, {slope_after3}")
```
